```python
"""
-------------------------------------------------------------------------
Name:            Joshua Navarro
Date:            March 6, 2025
Project #:       2
Status:          Completed
Class:           PHYS 2425
-------------------------------------------------------------------------
Objective:  
This project models the fall of a soccer ball from a height of 200 meters, 
taking into account both gravity and air resistance. The simulation tracks 
the ball’s height, velocity, and acceleration over time, illustrating how 
air drag affects its motion.
-------------------------------------------------------------------------
"""
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = -9.81  # Gravity (m/s^2), negative since downward
m = 0.425  # Mass of the ball in kg
y = 200  # Initial height (m)
v = 0  # Initial velocity (m/s)
t = 0  # Initial time (s)
dt = 0.01  # Time increment (s)
r = 0.115  # Radius of the ball (m)
pi_value = np.pi
C = 0.25  # Drag coefficient
p = 1.2  # Air density (kg/m^3)
A = pi_value * r**2  # Cross-sectional area (m^2)

# terminal velocity
v_term = -np.sqrt((2 * m * abs(g)) / (C * A * p))

# Lists to store values
yValues = []
vValues = []
tValues = []
aValues = []

# Simulation loop
while y > 0:
    yValues.append(y)
    vValues.append(v)
    tValues.append(t)

    # Drag force (always opposing motion)
    Fdrag = (0.5 * C * A * p * v**2) * (-np.sign(v))  # Ensures opposite direction to velocity
    Fnet = m * g + Fdrag  # Net force

    #acceleration
    a = Fnet / m
    aValues.append(a)

    # Update velocity and position
    v = v + a * dt
    y = y + v * dt
    t = t + dt

# Plot results
plt.figure(figsize=(8, 5))
plt.grid()
plt.plot(tValues, yValues, label="Height (y) vs Time (t)")
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.grid()
plt.plot(tValues, vValues, label="Velocity (v) vs Time (t)")
plt.axhline(y=v_term, color='r', linestyle='--', label="Terminal Velocity")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.grid()
plt.plot(tValues, aValues, label="Acceleration (a) vs Time (t)")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s²)")
plt.legend()
plt.show()
```
